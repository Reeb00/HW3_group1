```python
merged_df_head = merged_df.loc[:10,:]
merged_df_head
```

```python
import geopandas as gpd
from shapely.geometry import Point
import folium

# Create a GeoDataFrame
gdf = gpd.GeoDataFrame(merged_df_head, geometry=gpd.points_from_xy(merged_df_head.lng, merged_df_head.lat))

# Create a Folium map centered at the mean location of the universities
m = folium.Map(location=[merged_df_head['lat'].mean(), merged_df_head['lng'].mean()], zoom_start=13)

# Add a marker for each university
for index, row in gdf.iterrows():
   # print(index)
   # print(row)
   folium.Marker([row['lat'], row['lng']], popup=row['universityName']).add_to(m)

# Show the map
```
